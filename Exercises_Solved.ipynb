# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
df = pd.read_csv('dataset/company_sales_data.csv')
print(df.head())
print(df.columns.tolist())
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](img/01.png)

```python
monthList = df['month_number'].tolist()
profitList = df['total_profit'].tolist()

plt.figure(figsize=(10, 5))
plt.plot(monthList, profitList)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Company Sales Data')
plt.xticks(monthList)
plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](img/02.png)

```python
monthList = df['month_number'].tolist()
profitList = df['total_profit'].tolist()

plt.figure(figsize=(10, 5))
plt.plot(monthList, profitList,
         label='Profit data of last year',
         color='r',
         marker='o',
         markerfacecolor='k',
         linestyle='--',
         linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.title('Company profit per month')
plt.xticks(monthList)
plt.yticks([100000, 150000, 200000, 250000, 300000, 350000, 400000, 450000])
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

### Exercise 3: Read all product sales data and show it using a multiline plot

![img](img/03.png)

```python
monthList = df['month_number'].tolist()

plt.figure(figsize=(10, 5))
plt.plot(monthList, df['facecream'], label='Face cream', marker='o')
plt.plot(monthList, df['facewash'], label='Face wash', marker='o')
plt.plot(monthList, df['toothpaste'], label='Toothpaste', marker='o')
plt.plot(monthList, df['bathingsoap'], label='Bathing soap', marker='o')
plt.plot(monthList, df['shampoo'], label='Shampoo', marker='o')
plt.plot(monthList, df['moisturizer'], label='Moisturizer', marker='o')

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Sales data')
plt.xticks(monthList)
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](img/04.png)

```python
monthList = df['month_number'].tolist()
toothpasteList = df['toothpaste'].tolist()

plt.figure(figsize=(10, 5))
plt.scatter(monthList, toothpasteList, label='Toothpaste sales data')
plt.xlabel('Month Number')
plt.ylabel('Number of units sold')
plt.title('Toothpaste sales data')
plt.xticks(monthList)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](img/05.png)

```python
monthList = df['month_number'].tolist()

plt.figure(figsize=(10, 5))
x = range(len(monthList))
width = 0.4

plt.bar([i - width/2 for i in x], df['facecream'], width=width, label='Face cream')
plt.bar([i + width/2 for i in x], df['facewash'], width=width, label='Face wash')

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Face cream and face wash sales data')
plt.xticks(list(x), monthList)
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

### Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

![img](img/06.png)

```python
monthList = df['month_number'].tolist()
bathingsoapList = df['bathingsoap'].tolist()

plt.figure(figsize=(10, 5))
plt.bar(monthList, bathingsoapList, label='Bathingsoap')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Bathing soap sales data')
plt.xticks(monthList)
plt.legend(loc='upper left')
plt.grid(True)
plt.savefig('bathing_soap_sales.png', dpi=100, bbox_inches='tight')
plt.show()
print("Plot saved as 'bathing_soap_sales.png'")
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](img/07.png)

```python
profitList = df['total_profit'].tolist()

plt.figure(figsize=(10, 5))
plt.hist(profitList, bins=5, rwidth=0.9)
plt.xlabel('Total profit in range')
plt.ylabel('Frequency')
plt.title('Profit data')
plt.grid(True)
plt.show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](img/08.png)

```python
labels = ['FaceCream', 'FaceWash', 'ToothPaste', 'BathingSoap', 'Shampoo', 'Moisturizer']
totals = [
    df['facecream'].sum(),
    df['facewash'].sum(),
    df['toothpaste'].sum(),
    df['bathingsoap'].sum(),
    df['shampoo'].sum(),
    df['moisturizer'].sum()
]

plt.figure(figsize=(8, 8))
plt.pie(totals, labels=labels, autopct='%1.1f%%')
plt.title('Sales data')
plt.legend(loc='lower right')
plt.show()
```

### Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

![img](img/09.png)

```python
monthList = df['month_number'].tolist()

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

ax1.plot(monthList, df['bathingsoap'], color='b', marker='o', linewidth=2, label='Bathing soap')
ax1.set_xlabel('Month Number')
ax1.set_ylabel('Sales units in number')
ax1.set_title('Sales data of Bathing soap')
ax1.legend(loc='upper left')
ax1.set_xticks(monthList)
ax1.grid(True)

ax2.plot(monthList, df['facewash'], color='r', marker='o', linewidth=2, label='Face wash')
ax2.set_xlabel('Month Number')
ax2.set_ylabel('Sales units in number')
ax2.set_title('Sales data of Face wash')
ax2.legend(loc='upper left')
ax2.set_xticks(monthList)
ax2.grid(True)

plt.tight_layout()
plt.show()
```

### Exercise 10: Read all product sales data and show it using the stack plot

![img](img/10.png)

```python
monthList = df['month_number'].tolist()

plt.figure(figsize=(10, 5))
plt.stackplot(monthList,
              df['facecream'],
              df['facewash'],
              df['toothpaste'],
              df['bathingsoap'],
              df['shampoo'],
              df['moisturizer'],
              labels=['FaceCream', 'FaceWash', 'ToothPaste', 'BathingSoap', 'Shampoo', 'Moisturizer'])

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('All product sales data using stack plot')
plt.xticks(monthList)
plt.legend(loc='upper left')
plt.show()
```
