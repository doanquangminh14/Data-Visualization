```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = 'dataset/7_OneCatOneNum.csv'
data = pd.read_csv(path)
df = data.sort_values(by = ['Value'], ascending=False)
df
```

```python
df.isna().sum()
```

Xử lý các quốc gia 

```python
df['Value'] = df['Value'].fillna(0)
df
```

Bar Chart

```python
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.set_style("whitegrid")
ax = sns.barplot(x='Value', y='Country', data=df, palette='viridis')
plt.title('Top Countries by Value', fontsize=16, fontweight='bold')
plt.xlabel('Value', fontsize=12)
plt.ylabel('Country', fontsize=12)
for i in ax.containers:
    ax.bar_label(i, padding=5)
plt.tight_layout()
plt.show()
```

Lolipop Chart

```python
plt.figure(figsize=(10, 8))
my_range = range(1, len(df.index) + 1)
plt.hlines(y=df['Country'], xmin=0, xmax=df['Value'], color='skyblue', alpha=0.5, linewidth=3)
plt.plot(df['Value'], df['Country'], "o", markersize=10, color='blue', alpha=0.8)
for i, val in enumerate(df['Value']):
    plt.text(val + 100, i, f'{int(val)}', va='center', fontsize=10)
plt.title('Lollipop Chart: Top Countries by Value', loc='left', fontsize=16, fontweight='bold')
plt.xlabel('Value')
plt.ylabel('Country')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
