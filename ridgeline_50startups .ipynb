```python
import pandas as pd
import joypy
import matplotlib.pyplot as plt
import matplotlib.cm as cm

df = pd.read_csv('dataset/50-startups.csv')
print(df.shape)
df.head()
```

##  Ridgeline Chart tổng thể (toàn bộ các biến số)

```python

numeric_cols = ['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']

fig, axes = joypy.joyplot(
    df[numeric_cols],
    figsize=(10, 6),
    colormap=cm.Set2,
    alpha=0.7,
    title='Ridgeline Chart - Phân bố tổng thể các biến số'
)

plt.xlabel('Giá trị')
plt.tight_layout()
plt.show()
```

### Đánh giá hình tổng thể

- **R&D Spend & Marketing Spend**: Phân bố lệch phải (right-skewed), nhiều công ty chi tiêu ở mức vừa phải, một số ít chi rất cao.
- **Administration**: Phân bố tương đối tập trung, ít biến động hơn so với các chi phí khác.
- **Profit**: Phân bố trải rộng, có xu hướng lệch phải, một số startup đạt lợi nhuận rất cao.
- Nhìn chung, các biến chi tiêu và lợi nhuận không phân bố chuẩn (normal distribution).

##  Ridgeline Chart theo nhóm cột State

###  Phân bố Profit theo State

```python
fig, axes = joypy.joyplot(
    df,
    by='State',
    column='Profit',
    figsize=(10, 5),
    colormap=cm.Pastel1,
    alpha=0.8,
    title='Ridgeline Chart - Phân bố Profit theo State'
)

plt.xlabel('Profit')
plt.tight_layout()
plt.show()
```

###  Phân bố R&D Spend theo State

```python
fig, axes = joypy.joyplot(
    df,
    by='State',
    column='R&D Spend',
    figsize=(10, 5),
    colormap=cm.Set1,
    alpha=0.8,
    title='Ridgeline Chart - Phân bố R&D Spend theo State'
)

plt.xlabel('R&D Spend')
plt.tight_layout()
plt.show()
```

###  Phân bố Marketing Spend theo State

```python
fig, axes = joypy.joyplot(
    df,
    by='State',
    column='Marketing Spend',
    figsize=(10, 5),
    colormap=cm.Accent,
    alpha=0.8,
    title='Ridgeline Chart - Phân bố Marketing Spend theo State'
)

plt.xlabel('Marketing Spend')
plt.tight_layout()
plt.show()
```

## Đánh giá hình theo nhóm State

### Profit theo State:
- **New York** và **California** có phân bố Profit khá tương đồng nhau, trải rộng và có đỉnh (peak) ở mức cao.
- **Florida** có phân bố tập trung hơn, đỉnh thấp hơn một chút so với hai bang còn lại.
- Cả 3 bang đều có phân bố lệch phải, cho thấy phần lớn startup đạt lợi nhuận ở mức trung bình, chỉ một số ít đạt lợi nhuận rất cao.

### R&D Spend theo State:
- Ba bang có hình dạng phân bố khá giống nhau (tương đồng về mức chi R&D).
- Không có sự khác biệt rõ ràng giữa các bang trong chi tiêu R&D, cho thấy mức đầu tư R&D không phụ thuộc nhiều vào địa điểm.

### Marketing Spend theo State:
- Phân bố Marketing Spend có sự chênh lệch nhất giữa các bang.
- **California** có xu hướng chi marketing cao hơn, phân bố trải rộng sang phải.
- **Florida** có phân bố tập trung ở mức thấp hơn.

### Nhận xét chung:
- Ridgeline chart cho phép so sánh trực quan phân bố giữa 3 nhóm State một cách hiệu quả.
- Sự chồng lấp giữa các đường phân bố cho thấy 3 bang có đặc điểm kinh doanh tương đối giống nhau.
- Biến **Marketing Spend** là biến có sự khác biệt rõ nhất giữa các bang.
