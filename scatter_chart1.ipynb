```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('dataset/toyota.csv')

```

```python
df.isnull().sum()
```

```python
fig, axes = plt.subplots(3, 2, figsize=(15, 18))
sns.set_style("whitegrid")


sns.scatterplot(data=df, x='mileage', y='price', ax=axes[0, 0], alpha=0.5)
axes[0, 0].set_title('Price vs Mileage')


sns.scatterplot(data=df, x='tax', y='price', ax=axes[0, 1], alpha=0.5)
axes[0, 1].set_title('Price vs Tax')


sns.scatterplot(data=df, x='mpg', y='price', ax=axes[1, 0], alpha=0.5)
axes[1, 0].set_title('Price vs MPG')


sns.scatterplot(data=df, x='engineSize', y='price', ax=axes[1, 1], alpha=0.5)
axes[1, 1].set_title('Price vs EngineSize')


sns.scatterplot(data=df, x='engineSize', y='mpg', ax=axes[2, 0], alpha=0.5)
axes[2, 0].set_title('MPG vs EngineSize')


sns.scatterplot(data=df, x='mpg', y='mileage', ax=axes[2, 1], alpha=0.5)
axes[2, 1].set_title('Mileage vs MPG')

plt.tight_layout()
plt.show()
```

1. Price và Mileage (Giá xe và Số dặm đã đi):

Xu hướng: Tương quan âm. Xe có số dặm (mileage) càng cao thì giá (price) càng giảm.

Phân tán: Các điểm dữ liệu tập trung dày đặc ở mức số dặm thấp (dưới 50,000) với mức giá dao động mạnh. Khi số dặm lớn (trên 100,000), giá xe có xu hướng hội tụ về một dải hẹp ở mức thấp.

2. Price và Tax (Giá xe và Thuế):

Xu hướng: Dữ liệu tụ thành các dải dọc (cụm). Điều này do thuế xe hơi thường được phân chia thành các mức cố định (ví dụ: 0, 145, 150...) chứ không phải là một giá trị liên tục.

Quan hệ: Những xe có giá rất cao thường chịu mức thuế cao, tuy nhiên vẫn có nhiều xe giá rẻ/trung bình nằm ở cùng nhóm thuế với xe giá cao.

3. Price và MPG (Giá xe và Mức tiêu thụ nhiên liệu - dặm/gallon):

Xu hướng: Không có tương quan rõ rệt. Dữ liệu tập trung thành một đám mây lớn ở vùng MPG từ 40-70.

Đặc điểm: Một số xe có mức MPG rất cao (tiết kiệm nhiên liệu, có thể là dòng hybrid) nhưng giá vẫn nằm ở mức trung bình hoặc thấp.

4. Price và EngineSize (Giá xe và Dung tích động cơ):

Xu hướng: Tương quan dương rõ rệt. Dung tích động cơ càng lớn thì giá xe càng cao.

Đặc điểm: Tương tự như Tax, Engine Size cũng phân bố theo các giá trị rời rạc (1.0, 1.5, 2.0...). Các cụm động cơ lớn (trên 2.5) có giá vượt trội hẳn so với động cơ nhỏ (dưới 2.0).

5. MPG và EngineSize (Tiêu thụ nhiên liệu và Dung tích động cơ):

Xu hướng: Tương quan âm nhẹ. Dung tích động cơ lớn thường có xu hướng chỉ số MPG thấp hơn (tốn nhiên liệu hơn).

Phân tán: Ở dung tích động cơ nhỏ (đặc biệt là 1.0 đến 1.8), biên độ dao động của MPG rất lớn, cho thấy công nghệ động cơ hoặc thiết kế có ảnh hưởng mạnh đến hiệu suất.

6. Mileage và MPG (Số dặm đã đi và Tiêu thụ nhiên liệu):

Xu hướng: Không có mối tương quan nào đáng kể. Hai biến số này hoàn toàn độc lập với nhau.

Phân tán: Sự phân bố lan rộng ngẫu nhiên, cho thấy số dặm một chiếc xe đã chạy không quyết định hoặc ảnh hưởng tới thông số thiết kế tiêu thụ nhiên liệu của nhà sản xuất.
